```python
print("The data is about hours spent talking on cellphone for males and females at Midwestern university")
print("It is desired to test with 2 hypothesis:")
print("H0: u1==u2 and H1: u1!=u2 judging the mean from males for u1 and females for u2")

males = [
    12, 4, 11, 13, 11, 9, 9, 10, 10, 7, 7, 12, 8, 9, 15, 10, 11, 12, 7, 8, 8, 9, 11, 10, 9, 
    10, 9, 9, 7, 9, 11, 7, 10, 10, 12, 9, 12, 12, 8, 13, 9, 10, 8, 11, 10, 13, 13, 9, 10, 13
]

females = [
    11, 9, 7, 10, 9, 10, 10, 7, 9, 10, 11, 8, 9, 6, 11, 10, 7, 9, 12, 14, 11, 12, 12, 8, 12, 
    12, 9, 10, 11, 7, 12, 7, 9, 9, 11, 10, 8, 13, 8, 10, 9, 9, 9, 11, 9, 9, 8, 9, 12, 11
]

entire_dataset = males + females

def calculate_stats(data, title):
    n = len(data)
    mean = sum(data) / n
    sorted_data = sorted(data)
    
    mid = n // 2
    median = (sorted_data[mid-1] + sorted_data[mid]) / 2 if n % 2 == 0 else sorted_data[mid]
    
    variance = sum((x - mean)**2 for x in data) / (n - 1)
    std_dev = variance**0.5
    
    skew = (n / ((n-1)*(n-2))) * sum(((x-mean)/std_dev)**3 for x in data)
    
    kurt_prefix = (n * (n + 1)) / ((n - 1) * (n - 2) * (n - 3))
    kurt_sum = sum(((x - mean) / std_dev)**4 for x in data)
    kurt_adj = (3 * (n - 1)**2) / ((n - 2) * (n - 3))
    kurtosis = (kurt_prefix * kurt_sum) - kurt_adj

    def get_percentile(p):
        idx = (n - 1) * p
        low = int(idx)
        high = low + 1
        if high < n:
            return sorted_data[low] + (idx - low) * (sorted_data[high] - sorted_data[low])
        return sorted_data[low]

    margin = 1.96 * (std_dev / (n**0.5))

    print(f"--- {title} ---")
    print(f"Mean: {mean:.2f}")
    print(f"Median: {median:.2f}")
    print(f"Standard Deviation: {std_dev:.2f}")
    print(f"Skewness: {skew:.4f}")
    print(f"Kurtosis: {kurtosis:.4f}")
    print(f"Quartiles:")
    print(f"Q1={get_percentile(0.25):.2f}")
    print(f"Q2={get_percentile(0.5):.2f}")
    print(f"Q3={get_percentile(0.75):.2f}")
    print(f"Deciles:")
    print(f"D1={get_percentile(0.1):.2f}")
    print(f"D2={get_percentile(0.2):.2f}")
    print(f"D3={get_percentile(0.3):.2f}")
    print(f"D4={get_percentile(0.4):.2f}")
    print(f"D5={get_percentile(0.5):.2f}")
    print(f"D6={get_percentile(0.6):.2f}")
    print(f"D7={get_percentile(0.7):.2f}")
    print(f"D8={get_percentile(0.8):.2f}")
    print(f"D9={get_percentile(0.9):.2f}")
    print(f"95% Confidence Interval for Mean: \n({mean - margin:.2f}, {mean + margin:.2f})\n")

calculate_stats(entire_dataset, "ENTIRE DATASET")
calculate_stats(males, "MALES")
calculate_stats(females, "FEMALES")

print("...")
print("-If we descibe about hours of talking on cellphone in a week of that university")
print("For the entirte dataset, college students spent almost 10 hours in average")
print("and its longer than it used to be")
print("-In testing with 2 hypothesis, this computation or solution proves H1 or Hypothesis 1 correct.")
print("-Comparing both genders by the dataset, males spent a bit longer than females in talking on cellphone time spent")
print("Its possible there are similar results yet same conclusion related to this topic outside university")
print("-For the confidence interval of 95%, for the entire dataset, it is likely between 9.44 and 10.20")
print("To compare males and females in the confidence interval, their difference is slight or just significant")
print("-For their pattern or differences, males are shown to have spent more time than females")
print("we may conclude that males spent a bit longer due to how open and willingness to go on to communicate personally through cellphones")
```
